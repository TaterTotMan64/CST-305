```python
import numpy as np
import matplotlib.pyplot as plt

# Define the recurrence relation
def recurrence_relation(n, a):
    if n == 0:
        return 0  # a_0 = 0
    elif n == 1:
        return 1  # a_1 = 1
    else:
        return -a[n-2] / (n * (n-1))

# Calculate coefficients up to n â‰¤ 8
a = [0] * 9  # Initialize coefficients list (0 to 8)
a[0] = 0  # Initial condition: a_0 = 0
a[1] = 1  # Initial condition: a_1 = 1

for n in range(2, 9):
    a[n] = recurrence_relation(n, a)

# Print coefficients
print("Coefficients up to n=8:")
for i in range(9):
    print(f"a_{i} = {a[i]}")

# Define the Taylor series up to x^8
def taylor_series(x, a):
    return sum(a[n] * x**n for n in range(9))

# Generate x values for plotting
x_vals = np.linspace(-2, 2, 400)
y_vals = taylor_series(x_vals, a)

# Plot the Taylor series
plt.plot(x_vals, y_vals, label="Taylor Series (n=8)", color='blue')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.title("Taylor Series Expansion up to n=8")
plt.xlabel("x")
plt.ylabel("y(x)")
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

# Define the differential equation
def diff_eq(y, x):
    y0, y1 = y
    dydx = [y1, (x - y0) / (x**2 + 4)]
    return dydx

# Initial conditions
y0 = [0, 0]  # y(0) = 0, y'(0) = 0 (assuming some initial conditions)

# Generate x values for the solution
x = np.linspace(0.1, 10, 400)  # Range for x values, avoiding x=0 to prevent division by zero

# Solve the differential equation
sol = odeint(diff_eq, y0, x)

# Extract solutions
y_sol = sol[:, 0]
y_prime_sol = sol[:, 1]

# Plot the solution
plt.plot(x, y_sol, label='y(x)')
plt.plot(x, y_prime_sol, label="y'(x)")
plt.xlabel('x')
plt.ylabel('y(x) and y\'(x)')
plt.title('Solution to the Differential Equation')
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt

def differential_equation(x, y):
    y, z = y
    dydx = z
    dzdx = (x - y) / (x**2 + 4)
    return [dydx, dzdx]

# Set initial conditions (adjust as needed)
initial_condition = [0, 0]  # Initial values for y and y'

# Define the interval of integration
x_span = [0, 10]  # Solve from x = 0 to x = 10

# Solve the equation
solution = solve_ivp(differential_equation, x_span, initial_condition)

# Extract the solution
x_values = solution.t
y_values = solution.y[0]

# Plot the solution
plt.plot(x_values, y_values)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Solution of the Differential Equation')
plt.grid(True)
plt.show()
```
