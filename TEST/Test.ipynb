```python
import matplotlib.pyplot as plt

# Simulate file fragments in a file system
file_fragments = [
    (1, 2), (5, 3), (0, 1), (8, 2), (3, 1), (7, 1),
    (12, 3), (9, 1), (15, 2), (18, 1), (21, 1)
]

# Extract positions and sizes
positions, sizes = zip(*file_fragments)

# Create a scatter plot
plt.figure(figsize=(10, 5))
plt.scatter(positions, sizes, s=[size*20 for size in sizes], alpha=0.6, edgecolors='w', linewidth=2)
plt.title('File Fragmentation Scatter Plot')
plt.xlabel('Fragment Start Position')
plt.ylabel('Fragment Size')
plt.show()
```
